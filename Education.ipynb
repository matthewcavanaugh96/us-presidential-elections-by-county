```python
values_to_keep = [
    "Percent of adults who are not high school graduates, 2000", 
"Percent of adults who are high school graduates (or equivalent), 2000", 
"Percent of adults completing some college or associate degree, 2000",
"Percent of adults with a bachelor's degree or higher, 2000",
"Percent of adults who are not high school graduates, 2008-12",              
"Percent of adults who are high school graduates (or equivalent), 2008-12",   
"Percent of adults completing some college or associate degree, 2008-12",    
"Percent of adults with a bachelor's degree or higher, 2008-12" 
]

education = education[education['Attribute'].isin(values_to_keep)]
```

Rename columns for processing

```python
education['Attribute'] = education['Attribute'].str.replace("Percent of adults who are not high school graduates, 2008-12", "Less than high school, 2008")
education['Attribute'] = education['Attribute'].str.replace("Percent of adults who are high school graduates (or equivalent), 2008-12", "High school graduate, 2008")
education['Attribute'] = education['Attribute'].str.replace("Percent of adults completing some college or associate degree, 2008-12", "At least some college, 2008")
education['Attribute'] = education['Attribute'].str.replace("Percent of adults with a bachelor's degree or higher, 2008-12", "Bachelor's degree or higher, 2008")
education['Attribute'] = education['Attribute'].str.replace("Percent of adults who are not high school graduates, 2000", "Less than high school, 2000")
education['Attribute'] = education['Attribute'].str.replace("Percent of adults who are high school graduates (or equivalent), 2000", "High school graduate, 2000")
education['Attribute'] = education['Attribute'].str.replace("Percent of adults completing some college or associate degree, 2000", "At least some college, 2000")
education['Attribute'] = education['Attribute'].str.replace("Percent of adults with a bachelor's degree or higher, 2000", "Bachelor's degree or higher, 2000")
```

```python
# Split into description and year
education[['Attribute', 'Year']] = education['Attribute'].str.extract(r'^(.*),\s*(\d{4})$')

education
```

```python
# Drop national and state aggregates, only keep county equivalents
education = education[education['Area name'].str.contains('County|City|city|Parish|District of Columbia', na=False)]
```

```python
# Drop Alaska
education = education[~education['State'].str.contains('AK', na=False)]
```

```python
education_pivoted = education.pivot_table(
    index=["FIPS Code", "Year", "State", "Area name"],
    # or just 'County' if there's no year
    columns="Attribute",
    values="Value"
).reset_index()

education_pivoted = education_pivoted.rename_axis(None, axis=1)
education_pivoted
```

```python
education_pivoted['Year'] = education_pivoted['Year'].astype(int)
```

```python
# Create a DataFrame for 2004 using 2000 values
rows_2004 = education_pivoted[education_pivoted['Year'] == 2000]
# Change the Year to 2004
rows_2004['Year'] = 2004

# Create a DataFrame for 2012 using 2008 values
rows_2012 = education_pivoted[education_pivoted['Year'] == 2008]
# Change the Year to 2012
rows_2012['Year'] = 2012

# Create a DataFrame for 2016 using 2008 values
rows_2016 = education_pivoted[education_pivoted['Year'] == 2008]
# Change the Year to 2016
rows_2016['Year'] = 2016


# Append to original DataFrame
education_pivoted = pd.concat([education_pivoted, rows_2004, rows_2012, rows_2016], ignore_index=True)
```

```python
education_pivoted
```

```python
education_pivoted['High school graduate'] = education_pivoted['High school graduate'] + education_pivoted['At least some college'] + education_pivoted["Bachelor's degree or higher"]
education_pivoted['At least some college'] = education_pivoted['At least some college'] + education_pivoted["Bachelor's degree or higher"]
```
