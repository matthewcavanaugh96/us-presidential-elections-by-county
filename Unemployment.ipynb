```python
import pandas as pd
```

```python
unemployment = pd.read_csv("/Users/matthewcavanaugh/Desktop/Data analysis work/Election 2000 2020/estimates/Unemployment2023.csv")
unemployment
```

```python
# Split attribute into 'Attribute' and 'Year'
unemployment[['Attribute', 'Year']] = unemployment['Attribute'].str.extract(r'(.+?)_(\d{4})')

# Convert Year to integer (optional but recommended for sorting/filtering)
unemployment['Year'] = unemployment['Year'].astype(int)
```

```python
# Drop national and state aggregates, only keep county equivalents
unemployment = unemployment[unemployment['Area_Name'].str.contains('County|City|city|Parish|District of Columbia', na=False)]
```

```python
# Remove ", XX" at the end of the string
unemployment['Area_Name'] = unemployment['Area_Name'].str.replace(r',\s*[A-Z]{2}$', '', regex=True)
```

```python
unemployment = unemployment[unemployment['Attribute'].str.contains('Unemployment_rate', na=False)]
```

```python
unemployment.rename(columns={'Value': 'Unemployment rate'}, inplace=True)
unemployment.drop('Attribute', axis=1, inplace=True)
unemployment
```
